```
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn import linear_model
import statsmodels.api as sm
```

```
df =  pd.read_csv('US.csv', index_col = 0)
df.drop(['Republican','Democrat','Neutral'], inplace = True, axis = 1)
```

```
df.head()
```

```
lm = linear_model.LinearRegression()
lm.fit(df[['Income','Education','Religion']],df.Abortion)
lm.coef_
```

```
df_dummy = pd.get_dummies(df['Party'])
df_dummy = pd.concat([df,df_dummy],axis = 1)
df_dummy.drop(['Party','Democrat'], axis = 1, inplace = True)
df_dummy
```

```
model = sm.OLS.from_formula('Abortion ~ Income + Education + Religion', df).fit()
model.summary()




```

```
model_2 = sm.OLS(df_dummy['Abortion'], sm.add_constant(df_dummy[['Income','Education','Religion','Republican']])).fit()

model_2.summary()
```
