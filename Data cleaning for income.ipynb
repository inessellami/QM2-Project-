```
import pandas as pd
```

```
income = pd.read_csv('income.csv')
income.head()
```

```
income.tail()
```

```
income.columns
```

```
clean1 = income.drop('2018', axis=1)
clean2 = clean1.drop('2017 (40)', axis=1)
clean3 = clean2.drop('2017', axis=1)
clean4 = clean3.drop('2016', axis=1)
clean5 = clean4.drop('2015', axis=1)
clean6 = clean5.drop('2013 (39)', axis=1)
clean7 = clean6.drop('2013 (38)', axis=1)
clean8 = clean7.drop('2012', axis=1)
clean9 = clean8.drop('2011', axis=1)
clean10 = clean9.drop('2010 (37)', axis=1)
clean11 = clean10.drop('2009 (36)', axis=1)
clean11.head()
```

```
clean11.drop(clean11.columns[clean11.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
clean11.head()
```

```
clean12 = clean11.drop(clean11.index[0])
clean12.head()
```

```
clean13 = clean12.drop(clean12.index[1])
clean13.head()
```

```
finalcleaned = clean13.rename(columns={'2014':'Median income in 2014'})
finalcleaned.head()
```

```
finalcleaned.to_csv(r'sample_data\Median Income.csv', index = False)
```
